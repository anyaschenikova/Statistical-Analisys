```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Считаем данные из файла

```python
xdata = pd.read_csv("xdata.csv", sep = ";", header = 0, index_col = 0, parse_dates = True) #read xdata
xdata.head()
```

### Выделим из всей таблицы необходимые колонки

```python
my_number = 13
data = xdata.iloc[:, my_number * 10: (my_number + 1) * 10] #cut nesessary columns
data
```

## Pассмотрим данные подробнее:

```python
data.describe()
```

### Построим графики сравнения максимальных, средних, и минимальных значений данных по колонкам

```python
fig, axs = plt.subplots(figsize = (7, 5))
plt.bar(range(data.shape[1]), data.max(axis=0), color = 'red', label = 'max')
plt.bar(range(data.shape[1]), data.mean(axis=0), color = 'blue', label = 'mean')
plt.bar(range(data.shape[1]), data.min(axis=0), color = 'green', label = 'min')
axs.legend()
axs.set_ylim([0,300000])
plt.show()
```

```python
fig, axs = plt.subplots(figsize = (7, 5))
plt.plot(range(data.shape[1]), data.max(axis=0), color = 'red', label = 'max')
plt.plot(range(data.shape[1]), data.mean(axis=0), color = 'blue', label = 'mean')
plt.plot(range(data.shape[1]), data.min(axis=0), color = 'green', label = 'min')
axs.legend()
axs.set_ylim([0,300000])
plt.show()
```

### Вывод:

Присутствуют пустые колонки на четных индексах. Максимальное, среднее и минимальное значения - наибольшие во второй колонке (1-ый индекс), в то время как оставшиеся четыре непустых колонки практически в 4 раза меньше второй колонки.

### Построим распределение данных в каждом столбце:

```python
for coloumn in data.columns:
    plt.figure(figsize=(5, 5))
    plt.hist(data[coloumn], bins=10, color = 'blue')
    plt.xlabel('Значения в колонке ' + coloumn)
    plt.ylabel('Распределение')
    plt.tight_layout()
```

```python
for coloumn in data.columns:
    fig, axs = plt.subplots(figsize = (8, 6))
    plt.scatter(range(data.shape[0]), data[coloumn], color = 'orange', label = coloumn)
    plt.ylabel('Значения в колонке ' + coloumn)
    plt.xlabel('Распределение')
    axs.legend()
    plt.show()
```

### Вывод:

В пустых столбцах все возможные значения - это нуль, поэтому распределение выглядит таким образом. В непустых столбцах данные шумные.

# Построим регрессию для data на ts.csv:

```python
target = pd.read_csv("ts.csv", sep = ";", header = 0, index_col = 0, parse_dates=True)
target.head()
```

### Распределение data на target:

```python
for coloumn in data.columns:
    plt.figure(figsize=(7, 5))
    plt.scatter(data[coloumn], target, color = 'yellow')
    plt.ylabel('target', size=15)
    plt.xlabel(coloumn, size=15)
    plt.tight_layout()
```

### Линейная регрессия для data на target:

```python
import sklearn.model_selection as sk
X_train, X_test, y_train, y_test = sk.train_test_split(data, target['ts'] , test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train, y_train)
predicted = clf.predict(X_test)
expected = y_test

from sklearn.metrics import r2_score
from sklearn.metrics import average_precision_score
from sklearn.metrics import mean_absolute_error
MAE = sum(abs(predicted - expected)) / len(y_test)
MSQERR = sum((predicted - expected)**2) / len(y_test)

plt.figure(figsize=(7, 5))
plt.scatter(expected, predicted, color = 'blue')
plt.plot([0, 0.6], [0, 0.6], '--k', color = 'black')
plt.axis('tight')
plt.xlabel('True value')
plt.ylabel('Predicted value')
plt.tight_layout()
```

### Вывод: 

Предсказанные с помощью линейной модели данные значительно меньше ожидаемых данных, значит модель сработала, но не оптимально, т.к. коеффициент ошибки велик

### Регрессия с помощью метода градиентного спуска для data на target:

```python
from sklearn.ensemble import GradientBoostingRegressor
clf1 = GradientBoostingRegressor()
clf1.fit(X_train, y_train)
predicted = clf1.predict(X_test)
expected = y_test


plt.figure(figsize=(7, 5))
plt.scatter(expected, predicted, color = "green")
plt.plot([0, 0.6], [0, 0.6], '--k', color = "blue")
plt.axis('tight')
plt.xlabel('True value')
plt.ylabel('Predicted value')
plt.tight_layout()
```

### Вывод: 

Метод градиентного спуска в данном случае с конкретными данными сработал лучше, так как предсказанные значения достаточно близки к ожидаемым значениям, а значит значение ошибки меньше
